# US STATES SCRAPING

```python
# Import of libraries

import pandas as pd 
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
import re
import difflib
```

```python
# Driver configuration

driver = 'chromedriver'
driver = webdriver.Chrome()
```

```python
# URL and call 

url = 'https://es.wikipedia.org/wiki/Anexo:Estados_y_territorios_de_los_Estados_Unidos'
driver.get(url)
```

```python
# Extraction of the whole table

wiki_table = driver.find_element(By.XPATH, '//*[@id="mw-content-text"]/div[1]/table[1]')
wiki_table
```

```python
# Extraction of data

us_states = wiki_table.text.split('\n')

us_states = [e.split() for e in us_states]

us_states
```

```python
# Creation and display of the data frame

us_states = pd.DataFrame(us_states)
us_states.head()
```

Deletion of rows and columns

```python
# Deletion of columns from index 1-18 (axis 1)

lst = list(range(1, 19))
us_states = us_states.drop(lst, axis = 1)
us_states.head()
```

```python
states_names = [
    'Alabama', 'Alaska', 'Arizona', 'Arkansas', 'California', 'Carolina del Norte',
    'Carolina del Sur', 'Colorado', 'Connecticut', 'Dakota del Norte', 'Dakota del Sur',
    'Delaware', 'Florida', 'Georgia', 'Hawái', 'Idaho', 'Illinois', 'Indiana', 'Iowa',
    'Kansas', 'Kentucky', 'Luisiana', 'Maine', 'Maryland', 'Massachusetts', 'Míchigan',
    'Minnesota', 'Misisipi', 'Misuri', 'Montana', 'Nebraska', 'Nevada', 'Nueva Hampshire',
    'Nueva Jersey', 'Nueva York', 'Nuevo México', 'Ohio', 'Oklahoma', 'Oregón', 'Pensilvania',
    'Rhode Island', 'Tennessee', 'Texas', 'Utah', 'Vermont', 'Virginia', 'Virginia Occidental',
    'Washington', 'Wisconsin', 'Wyoming'
]
```

```python
# Deletion of the rows to match number of states

us_states = us_states.drop([0, 1, 2, 3, 4])
us_states = us_states.reset_index(drop = True)
```

```python
# Changing the name of the column

us_states = us_states.rename(columns={ 0: 'state_name' })
```

```python
# Changing de values of the column

us_states['state_name'] = states_names
us_states.head()
```

```python
# Extraction of population data from url: https://datosmacro.expansion.com/demografia/poblacion/usa-estados?anio=2018

population = pd.read_clipboard()
population.head()
```

```python
# Addition of data from name of column to body of the data frame and sorting the index

new_data = {'737.717': 737.717}
population.loc[-1] = new_data
population.index = population.index + 1
population = population.sort_index()
population.head()
```

```python
# Rename of the just changed column

population.rename(columns={'737.717': 'population'}, inplace=True)
population.head()
```

```python
# Putting together both tables

us = pd.concat([us_states, population], axis=1)
us.head()
```

```python
# Keeping just the relevant columns for the analysis

us = us.loc[:, ['state_name', 'population']]
us
```
