```python
universities["states_id"] = universities['abbreviation']
```

```python
# Getting unique values 

universities['abbreviation'].unique()
```

```python
# Deleting blank spaces

universities['abbreviation'] = universities['abbreviation'].apply(lambda x: x.strip())
```

```python
universities['abbreviation'].unique()
```

```python
# Adding to the new column the corresponding ids

universities['states_id'] = universities['abbreviation'].apply(lambda x: dictio[x] if x in dictio else dictio['WA'])
```

```python
# Display of the data frame 

universities.head(10)
```

```python
# Export of the data frame

universities.to_csv('universities_limpio.csv', index = False)
```

```python
# Creation of a dictionary with names and ids of universities

dict_uni = {name: index for index, name in universities[['id', 'uni_name']].values}
dict_uni
```

```python
# Adding to the new column the corresponding ids

teams_players['uni_id'] = teams_players['college'].apply(lambda x: dict_uni[x])
```
