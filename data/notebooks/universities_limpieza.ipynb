# US universities

```python
# Import of libraries

import pandas as pd
```

```python
# CSV import

universities = pd.read_csv('../../data/csv/universities.csv')
```

```python
# Display of the data frame

universities.head(10)
```

```python
# Deletion of non-relevant columns

universities = universities.drop(['In-state', 'Undergrad Enrollment', 'index'], axis = 1)
universities.head()
```

```python
# Rename of columns 

universities.rename(columns = {'Name': 'uni_name', 'Location': 'state', 'Rank': 'rank',
                               'Description': 'description', 'Tuition and fees': 'fees'}, inplace = True)
universities.head()
```

```python
# Separation of the name of the state and its abbreviation into two different columns

universities['abbreviation'] = universities['state'].str.split(',').str.get(1)
universities['state'] = universities['state'].str.split(',').str.get(0)
universities.head()
```

```python
universities['fees'] = universities['fees'].str.replace(',', '').str.replace('$', '')
```

```python
# Rearranging the order of the columns

new_order = ['uni_name', 'rank', 'state', 'abbreviation', 'description', 'fees']
universities = universities[new_order]
universities.head()
```

```python
# Adding 'other' as a new row

new_row = pd.DataFrame({
  'uni_name': ['other'],
  'rank': ['unknown'],
  'state': ['other'],
  'abbreviation': ['unknown'],
  'description': ['unknown'],
  'fees': ['unknown']
})

universities = universities.append(new_row, ignore_index=True)
```

```python
universities.head()
```

```python
# Exporting the data frame into a CSV 

universities.to_csv('universities_limpio.csv')
```
