# US teams and players

This data is from 2018

```python
# Import of libraries

import pandas as pd
from fuzzywuzzy import fuzz
from fuzzywuzzy import process
```

```python
# CSV import

teams_players = pd.read_csv('../../data/csv/teams_players.csv')
```

```python
# Display of the data frame

teams_players.head(10)
```

```python
# Deletion of non-relevant columns

teams_players = teams_players.drop(['index', 'ID', '#'], axis = 1)
teams_players.head()
```

```python
# Rename of columns 

teams_players.rename(columns = {'Player': 'player_name', 'Pos': 'position', 
                                'HT': 'height','WT': 'weight', 'Age': 'age', 
                                'Exp': 'experience', 'College': 'college', 
                                'Team': 'team'}, inplace = True)
teams_players.head()
```

```python
# Checking for rows with all Nan values, deleting them and checking results

teams_players.dropna(how = 'all', inplace=True)
teams_players.isna()
```

```python
# Checking for unique values in the column 'college'

teams_players['college'].unique()
```

```python
# 'universities_limpio.csv' import to compare columns with university names

universities_limpio = pd.read_csv('../../data/csv_limpios/universities_limpio.csv')
```

```python
# View of the column 'uni_name' out of the csv 'universities_limpio'

universities_limpio['uni_name']
```

```python
# Setting the same university name in 'college' from 'uni_name' if they're 75% alike

for index, row in teams_players.iterrows():
    try:
        match = process.extractOne(row['college'], universities_limpio['uni_name'])
        if match[1] >= 85:
            teams_players.at[index, 'college'] = match[0]
    except:
        # Manejar la excepción aquí
        teams_players['college'] = 'other'
```

```python
# Exporting the data frame into a CSV 

teams_players.to_csv('teams_players_limpio.csv')
```
